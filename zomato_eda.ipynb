```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
```

```python
ratings
```

observations
1. when ratings is 4.5 to 4.9 its indicate that the excellent
2. when ratings is 4.0 to 4.4 its indicate that the Very Good
3. when ratings is 3.5 to 3.9 its indicate that the Good
4. when ratings is 2.5 to 3.4 its indicate that the Average
5. when ratings is 1.8 to 2.4 its indicate that the Poor
6. when ratings is 0.0 then its indicate that the reamaining peaople is not rated.

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='rating count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','blue'])
```
